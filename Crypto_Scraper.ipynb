```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import xlsxwriter


# Main function for scraping wanted sort types from coinmarketcap into diffeerent worksheets.
def crypto_scraper_all(list_of_wanted_sorts):

    work_book = xlsxwriter.Workbook('Crypto_Scraper.xlsx')
    
    driver = webdriver.Chrome(ChromeDriverManager().install())
    
    driver.get('https://coinmarketcap.com')

    for item in list_of_wanted_sorts:

        if item[0].lower() == 'popular':

            scraper_inside(driver, work_book, item, 'Popular')

        elif item[0].lower() == 'seven day change':
            driver.execute_script(f"window.scrollTo(0, 400)")
            seven_day_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, './/*[@id="__next"]/div/div[1]/div[2]/div/div/div[5]/table/thead/tr/th[6]/div/div/div/p'))).click()
            
            scraper_inside(driver, work_book, item, 'SevenDaysChange')

        elif item[0].lower() == 'volume twenty four hours':
            driver.execute_script(f"window.scrollTo(0, 400)")
            volume_day_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, './/*[@id="__next"]/div/div[1]/div[2]/div/div/div[5]/table/thead/tr/th[8]/div/div[1]/p'))).click()

            scraper_inside(driver, work_book, item, 'Volume24Hours')

        else:
            print('Not available sorting type!')
        
    work_book.close()



#Inside loop for scraping cryptos of wanted sort type.
def scraper_inside(driver, work_book, item, sort_type):
            loop_condition = True
            work_sheet_popular = work_book.add_worksheet(f'{sort_type}_Coins')
            work_sheet_popular.write(0, 0, '#')
            work_sheet_popular.write(0, 1, 'Name')
            work_sheet_popular.write(0, 2, 'Price')
            work_sheet_popular.write(0, 3, 'Market Cap')
            work_sheet_popular.write(0, 4, 'Volume Twenty Four Hours')

            i = 1
            page_number = 1


            while loop_condition:
                y = 200
                total_height = int(driver.execute_script("return document.body.scrollHeight"))

                while y <= total_height - 2000:

                    driver.execute_script(f"window.scrollTo(0, {y})")

                    y = y + 200


                coins = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.TAG_NAME, 'tr')))[1:101]

                if len(coins) < int(item[1]):
                    
                    for coin in coins:

                        work_sheet_popular.write(i, 0, str(i))

                        coin_infos = coin.find_elements(By.CSS_SELECTOR, 'a[class="cmc-link"]')

                        coin_name_container = coin_infos[0].get_attribute('href').split('ies/')
                        coin_name = coin_name_container[1][:-1]
                        work_sheet_popular.write(i, 1, coin_name)

                        coin_price = coin_infos[1].find_element(By.TAG_NAME, 'span').text
                        work_sheet_popular.write(i, 2, coin_price)

                        coin_market_cap = coin.find_element(By.CSS_SELECTOR, 'span[class="sc-1ow4cwt-1 ieFnWP"]').text
                        work_sheet_popular.write(i, 3, coin_market_cap)

                        volume_twenty_four_hours = coin.find_element(By.CSS_SELECTOR, 'p[class="sc-1eb5slv-0 hykWbK font_weight_500"]').text
                        work_sheet_popular.write(i, 4, volume_twenty_four_hours)

                        i = i + 1
                    
                    page_number = page_number + 1
                    item[1] = int(item[1]) - 100

                    driver.get(f'https://coinmarketcap.com/?page={page_number}')

                elif len(coins) == int(item[1]):

                    for coin in coins:
    
                        work_sheet_popular.write(i, 0, str(i))

                        coin_infos = coin.find_elements(By.CSS_SELECTOR, 'a[class="cmc-link"]')

                        coin_name_container = coin_infos[0].get_attribute('href').split('ies/')
                        coin_name = coin_name_container[1][:-1]
                        work_sheet_popular.write(i, 1, coin_name)

                        coin_price = coin_infos[1].find_element(By.TAG_NAME, 'span').text
                        work_sheet_popular.write(i, 2, coin_price)

                        coin_market_cap = coin.find_element(By.CSS_SELECTOR, 'span[class="sc-1ow4cwt-1 ieFnWP"]').text
                        work_sheet_popular.write(i, 3, coin_market_cap)

                        volume_twenty_four_hours = coin.find_element(By.CSS_SELECTOR, 'p[class="sc-1eb5slv-0 hykWbK font_weight_500"]').text
                        work_sheet_popular.write(i, 4, volume_twenty_four_hours)

                        i = i + 1
                    
                    loop_condition = False
                
                elif len(coins) > int(item[1]):
                        for coin in coins[0: int(item[1])]:
        
                            work_sheet_popular.write(i, 0, str(i))

                            coin_infos = coin.find_elements(By.CSS_SELECTOR, 'a[class="cmc-link"]')

                            coin_name_container = coin_infos[0].get_attribute('href').split('ies/')
                            coin_name = coin_name_container[1][:-1]
                            work_sheet_popular.write(i, 1, coin_name)

                            coin_price = coin_infos[1].find_element(By.TAG_NAME, 'span').text
                            work_sheet_popular.write(i, 2, coin_price)

                            coin_market_cap = coin.find_element(By.CSS_SELECTOR, 'span[class="sc-1ow4cwt-1 ieFnWP"]').text
                            work_sheet_popular.write(i, 3, coin_market_cap)

                            volume_twenty_four_hours = coin.find_element(By.CSS_SELECTOR, 'p[class="sc-1eb5slv-0 hykWbK font_weight_500"]').text
                            work_sheet_popular.write(i, 4, volume_twenty_four_hours)

                            i = i + 1
                    
                        loop_condition = False


crypto_scraper_all([['volume twenty four hours', '168']])









```
