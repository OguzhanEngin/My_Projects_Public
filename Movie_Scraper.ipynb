```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random


# Defining a function for going to wanted genre page and printing a random movie.
def imdb_movie_scraper(genre_input, number, ratinglow, ratinghigh):

    #Sets up a driver.
    driver = webdriver.Chrome(ChromeDriverManager().install())

    driver.get('https://www.imdb.com/feature/genre/?ref_=nv_ch_gr')
    

    genres = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, 'div[class="image"]')))

    for genre in genres:
        genre_title = genre.find_element(By.CLASS_NAME, "pri_image")


        if genre_input == genre_title.get_attribute('title'):
            genre_link = genre.find_element(By.TAG_NAME, 'a')
            driver.get(genre_link.get_attribute('href'))


            wanted_movie_list = []

            movie_scraper(driver, number, wanted_movie_list, ratinglow, ratinghigh)

            x = random.randint(0, len(wanted_movie_list) - 1)

            print(wanted_movie_list[x])

            break

       

# Defining a function for scraping a list of movies of wanted genre.
def movie_scraper(driver, number, wanted_movie, ratinglow, ratinghigh):
    loop_condition = True
    page_number = 0

    while loop_condition:



        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

        movies = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, 'div[class="lister-item-content"]')))

        if len(movies) < number:

            for movie in movies:
                try:
                    rating = movie.find_element(By.TAG_NAME, 'strong').text


                    if ratinglow <= float(rating) <= ratinghigh:
                        
                        movie_name_date = movie.find_element(By.CSS_SELECTOR, 'h3[class="lister-item-header"]').text



                        wanted_movie.append((movie_name_date, rating))
                except:
                    pass
            
            if page_number < 1:
                next_page_link = driver.find_element(By.XPATH, './/*[@id="main"]/div/div[4]/a').get_attribute('href')

                driver.get(next_page_link)
            
            else:
                next_page_link = driver.find_element(By.XPATH, './/*[@id="main"]/div/div[4]/a[2]').get_attribute('href')

                driver.get(next_page_link)

            number = number - 50
            page_number = page_number + 1



        elif len(movies) == number:

            for movie in movies:
                try:
                    rating = movie.find_element(By.TAG_NAME, 'strong').text


                    if ratinglow <= float(rating) <= ratinghigh:
                        
                        movie_name_date = movie.find_element(By.CSS_SELECTOR, 'h3[class="lister-item-header"]').text



                        wanted_movie.append((movie_name_date, rating))                
                except:
                    pass
            loop_condition = False

        else:

            new_movies = movies[0: number - 1]

            for movie in new_movies:
                try:

                    rating = movie.find_element(By.TAG_NAME, 'strong').text


                    if ratinglow <= float(rating) <= ratinghigh:
                        
                        movie_name_date = movie.find_element(By.CSS_SELECTOR, 'h3[class="lister-item-header"]').text



                        wanted_movie.append((movie_name_date, rating))   
                except:
                    pass
            loop_condition = False


imdb_movie_scraper('Adventure', 260, 4.0, 7.9)
```
